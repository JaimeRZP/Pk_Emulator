```python
import pyccl as ccl
import numpy as np
import pylab as plt
import itertools

import pymaster as nmt
```

First we want to set up a grid of h and Omega_m spanning across our priors 

```python
w_cdm  = np.linspace(0, 1, 2)
h  = np.linspace(0.64,  0.82, 2)
```

```python
cosmo = ccl.Cosmology(Omega_c=0.27, Omega_b=0.049, h=0.67, sigma8=0.81, n_s=0.96)
```

```python
nk = 100 # No. of points in k
k_arr = np.logspace(-4.7,5,nk)
```

```python
for i in range(len(w_cdm)):
    cosmo.Omega_c=w_cdm[i]
    for j in range(len(h)):
        cosmo.h=h[j]
        output =ccl.power.linear_matter_power(cosmo, k_arr, 1)
        print(output) 
```
